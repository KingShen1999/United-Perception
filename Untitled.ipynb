```python
from collections import *
import matplotlib.pyplot as plt
import cv2

freqDict = defaultdict(int)
sizeDict = defaultdict(int)

with open('train.txt', 'r') as reader:
    line = reader.readline()
    
    while line != '':  # The EOF char is an empty string
        line = reader.readline()
        
        tmp = line.split(" ")

        try:
            image = cv2.imread('/work/u2127085/Pratical_DL/'+tmp[0])
            sizeDict[image.shape]+=1
            freqDict[tmp[1][:-1]]+=1
        except:
            print(tmp)
```

```python
sizeDict
```

```python
totalSize = []
for key in sizeDict.keys():
    totalSize.append([sizeDict[key],key])
totalSize.sort()
```

```python
totalSize = totalSize[-1::-1]
totalSize
```

```python
x_size = []
y_size = []
for value,key in totalSize:
    x_size.append(str(key))
    y_size.append(value)
    
print(len(x_size))
print(len(y_size))
```

```python
x_size.sort()
x_size
```

```python
total = []
for key in freqDict.keys():
    total.append([freqDict[key],key])
total.sort()
```

```python
total = total[-1::-1]
```

```python
x = []
y = []
for value,key in total:
    x.append(key)
    y.append(value)
```
